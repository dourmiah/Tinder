On doit convertir les valeurs discrètes (âge, race, etc...) en integer

```python
cols = ["positin1", "age", "race", "age_o", "race_o", "met", "met_o", "goal", 
        "date", "go_out", "career_c", "imprace", "imprelig", "field_cd", "exphappy",
        "like", "prob", "like_o", "prob_o","match_es"]

for col in cols:
    df[col] = df[col].fillna(0).astype("int64")
    
```

```python
cols = ["mn_sat", "tuition", "income"]


for col in cols:
    df[col] = df[col].str.replace(",","")
    df[col] = df[col].astype("float64")
```

Valeurs manquantes

```python
plt.figure(figsize=(20, 10))
sns.heatmap(df.isna(), cbar=False)
```

Observations : 
La plupart des valeurs manquantes sont dans les colonnes qui contiennent les évaluations du candidat. 
On décide de supprimer ces colonnes, car elles ne sont pas utiles à l'étude

On va conserver dans le dataset les colonnes avec moins de 70 % de valeurs manquantes

```python
df = df[df.columns[(df.isna().sum() / df.shape[0] * 100) < 70]]
```

Suppression des colonnes inutiles

```python

# Identité ("iid", "id", "idg", "pid")
df.drop(columns=["iid", "id", "idg", "pid"], inplace=True)

# Colonnes décision ("dec" et "dec_o") : Redondantes avec la colonne "match"
df.drop(columns=["dec", "dec_o"], inplace=True)

# Intérêts/hobbies : redondants avec la corrélation d'intérêts (int_corr)
todrop = ["sports", "tvsports", "exercise", "dining", "museums", "art", "hiking", "gaming", "clubbing", "reading", "tv", "theater", "movies", "concerts", "music", "shopping", "yoga"]
df.drop(columns=todrop, inplace=True)

#  Les données remontées par le candidat suite au Speed Dating (rencontres, appels, etc...) : On sort du cadre de l'étude. Target = "matchs"
#  Colonnes : "satis_2", "length", "numdat_2", "you_call", "them_call", "date_3" 
df.drop(columns=["satis_2", "length", "numdat_2", "you_call", "them_cal", "date_3"], inplace=True)

# Toutes les évaluations du (de la) candidat(e) (attrx_x, sincx_x, etc...): Elles ne sont pas utiles pour l'analyse
todrop = ["pf_o_att", "pf_o_sin", "pf_o_int", "pf_o_fun", "pf_o_amb", "pf_o_sha"] 
criteriastodrop = ["attr", "sinc", "intel", "fun", "amb", "shar"]

for crit in criteriastodrop:
          todrop.append(crit)
          todrop.append(f"{crit}_o")
          todrop.append(f"{crit}1_s")
          todrop.append(f"{crit}3_s")

          for i in range(1, 6):
                 todrop.append(f"{crit}{i}_1")
                 todrop.append(f"{crit}{i}_2")
                 todrop.append(f"{crit}{i}_3")
df.drop(columns=todrop, inplace=True, errors='ignore')
```
